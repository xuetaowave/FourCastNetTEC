```python
import numpy as np
import h5py
```

```python
data = h5py.File('autoregressive_predictions__vis.h5' , 'r')
data.keys()
```

```python
true = np.array(data['ground_truth'])
pred = np.array(data['predicted'])
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(12, 4), dpi=150)
plt.subplot(1, 2, 1)
plt.pcolor(true[0, 1, 0].T)
plt.colorbar()

plt.subplot(1, 2, 2)
plt.pcolor(pred[0, 1, 0].T)
plt.colorbar()
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(8, 8), dpi=150)
plt.pcolor(true[0, 0, 0]-pred[0, 0, 0])
plt.colorbar()
```

```python
print(pred.shape)
rmse = np.sqrt(np.mean((pred-true)**2, axis=(0, 1, 3, 4)))
np.mean(true), rmse
```
